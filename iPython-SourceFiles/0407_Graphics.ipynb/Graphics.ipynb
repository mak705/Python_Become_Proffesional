In this example we are going to go through some of the features of iPython. It's main advantage is that we can visualize out data and show it in a more user friendly way.

```python
import pandas as pd
d = {'Name':["Giants", "Partriots", "Seahawks", "Bears", "49ers"], 'Stadium Capacity':[82500,75000,70000,61500,65000], 'Titles':[8,3,1,1,5]}
df = pd.DataFrame(d)
df.head()
```

Let's plot the data above:

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(d['Stadium Capacity'],d['Titles'])
plt.title('Titles vs Stadium Capacity')
plt.xlabel('Stadium Capacity')
plt.xlabel('Titles')
```

iPython offers some more advanced visualization options like histograms:

```python
import numpy as np
rng = np.random.RandomState(10) #deterministic random data
a = np.hstack((rng.normal(size=1000),
              rng.normal(loc=5, scale=2, size=1000)))
plt.hist(a, bins='auto')
plt.title("Histogram random data example")
plt.show()
```

Now the plotting code:

```python
plt.bar(d['Stadium Capacity'], d['Titles'], linewidth=5,tick_label=d['Name'])
plt.title("Titles vs Stadium Capacity")
plt.show()
```
